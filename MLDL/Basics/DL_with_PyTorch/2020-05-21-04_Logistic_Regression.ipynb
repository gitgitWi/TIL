### 01.-02. Binary Classification, Sigmoid Function

$H(x) = f(Wx + b)$ -> sigmoid function

$H(x) = \frac{1}{1+e^{-(Wx+b)}}$

#### numpy를 통해 Sigmoid Function 구현

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

#### 01. W = 1, b = 0 인 그래프

```python
def sigmoid (x) : 
    return 1/(1+np.exp(-x))
```

```python
x = np.arange(-5.1, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y, 'g')
plt.plot([0,0], [1.0,0.0], ':')
plt.title('Sigmoid Function')
plt.show()
```

#### 02. W 값 변화에 따른 경사도 변화

```python
y1 = sigmoid(0.5*x)
y2 = sigmoid(x)
y3 = sigmoid(2*x)

plt.plot(x, y1, 'r', linestyle = '--')
plt.plot(x, y2, 'g')
plt.plot(x, y3, 'b')
plt.plot([0,0],[1.0,0.0],':')
plt.title('Sigmoid Function')
plt.show()
```

#### 03. b 값 변화에 따른 좌우 이동

```python
y1 = sigmoid(x+0.5)
y2 = sigmoid(x+1)
y3 = sigmoid(x+1.5)

plt.plot(x, y)
plt.plot(x, y1, 'r', linestyle = '--')
plt.plot(x, y2, 'g')
plt.plot(x, y3, 'b', linestyle = '--')
plt.plot([0,0], [1.0,0.0], ':')
plt.title('Sigmoid Function')
plt.show()
```

#### 04. classification by sigmoid function

- output >= 0.5 : True
- output <= 0.5 : False

### 03. Cost Function

- Sigmoid Function에서 비용 함수를 미분한 함수는 Non-convex 형태

- local minimum - global minimum

- Sigmoid는 실제값이 1일 때 예측값이 0에 가까워지면 오차가 무한대로 발산, 반대로 실제값 0일때 예측값 1에 가까워지면 오차가 무한대로 발산

- => Log 함수가 이러한 형태를 충족

- 1개 오차에 대한 cost function : $cost(H(x), y) = -[ylogH(x) + (1-y)log(1-H(x))]$
